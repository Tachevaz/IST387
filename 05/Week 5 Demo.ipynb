## Part 1: U.S. Senate Data (HW 5 Dataset)

**Step 1:** Load the raw JSON file:

```r
dataset <- getURL("https://ist387.s3.us-east-2.amazonaws.com/data/role.json")
```

```r
dataset    #Looks kind of messy, doesn't it?
```

**Step 2:** We can **parse** it to make it more stuctured and easier to read using the **fromJSON()** function from the **jsonlite** package (the **::** simply means ***“from this package, use this function”***)

```r
readlines <- jsonlite::fromJSON(dataset)
```

```r
readlines
```

**Step 3**: Now, let's save this somewhat tidier table into a much more familiar R object - a **dataframe**:

```r
df <- readlines$objects$person
```

**Step 4:** Describing the data. <br>
We can explore the contents of the **df** dataframe using the **head()**, **tail()**, or **View()** functions:

```r
head(df)
```

From the output of **df**, we can get an idea about the meaning of a row vs. column in this dataframe - each row appears to correspond to 1 U.S. senator, and each column - to an attribute of theirs.

Let's figure out the dimensions of our dataframe using the **dim()** function which will tell us how many rows and columns **df** has:

```r
dim(df)
```

**Step 5:** Data Analysis <br>

We can ask questions of our data. For example, we could ask R to count the number of senators whose first name is "Charles" by using a combination of **subsetting [ ]** and the **nrow()** function which counts the number of rows we are interested in:

```r
nrow(df[df$firstname=="Charles",]) #There are 2 senators whose first name is Charles
```

Using the same logic, we can count the number of male senators - all we need to modify is the condition inside the **[ ]**:

```r
nrow(df[df$gender=="male",])
```

Remember the **is.na()** function we used in HW 3? We can use it here too, to see how many senators **DO NOT have a Twitter account**:

```r
nrow(df[is.na(df$twitterid),])
```

How about reversing the condition to get the number of senators who **DO** have a Twitter account? All we need to do is **negate** our original condition with an **exclamation point (!)** in front of the function we want to negate - in this case, **is.na()**:

```r
nrow(df[!is.na(df$twitterid),])
```

We can even combine multiple conditions using **AND (&)** and **OR (|)** conditions. The following returns the **number of senators who are male AND have a Twitter account**:

```r
nrow(df[df$gender=="male" & !is.na(df$twitterid),])
```

Looking at the number of rows can be illuminating but it doesn't show us the full picture. To do that, we may need to save our **subsetting** results in a new dataframe: 

```r
twitterMen <- df[df$gender=="male" & !is.na(df$twitterid),]
twitterMen
```

Sometimes data needs additional cleaning. <br>

See what the **substr()** function does to the **birthday** column - the first parameter is the column we want the function applied to; the second - the position of the beginning character we want, and the last - the position of the last character we need. In this case, we want to retain the first 4 characters in each row of the birthday column.

```r
twitterMen$year <- substr(twitterMen$birthday,1,4)
twitterMen
```

We can now calculate different things such as the oldest and youngest senator in our dataframe, but before we do that, we need to convert it to **numeric** - right now, R considers this column to be a string of characters. 

```r
twitterMen$year <-as.numeric(twitterMen$year)
twitterMen
```

```r
mean(twitterMen$year)
```

Finally, let's plot senators' birth years in a histogram to get a sense of their distribution: 

```r
hist(twitterMen$year)
```

## Part 2: NYC CitiBike data (Lab Dataset)

Let's take a look at another JSON dataset updated in real time - CitiBike in New York City.

Just as before, we will load the raw JSON and then parse it:

```r
station_link <- 'https://gbfs.citibikenyc.com/gbfs/en/station_status.json'
apiOutput <- getURL(station_link) 
apiData <- fromJSON(apiOutput)
```

Inspect the **apiData** object R created – it looks much tidier than the raw, unparsed **apiOutput**, and yet we don’t need all that data. <br>

Instead, we’d like to focus on a particular sub-element of the hierarchical apiData object: a **child object** called **stations** which contains some variables stored in **key:value** pairs of interest to us such as the number of bikes available at each station or the number of available docks. 

Extract only the **stations** part of **apiData** and store it in a dataframe: 

```r
stationStatus <- apiData$data$stations
```

Why did we need more than one $? **Answer**: Because **stations** is nested in **data** which is nested in **apiData**. So far, we have only looked at variables, aka columns, with only a single parent – a dataframe.  <br>

**stations** has a ***parent*** (**data**) but also a ***“grandparent”*** (**apiData**) – it’s the third branch in this tree-like data structure. You can see this for yourself if you run **View(apiData)** or something similar.

So far, so good. However, if you inspect the resulting dataframe **stationStatus**, you will see there are a lot of columns with most values missing. Let’s get rid of them and focus on the columns with values. Let’s retain the following 6 columns by saving their names to a vector called **cols**:


```r
cols <- c('num_bikes_disabled','num_docks_disabled', 'station_id',
          'num_ebikes_available', 'num_bikes_available', 'num_docks_available')
```

How does the following line of code accomplish our goal of updating the **stationStatus** dataframe to only contain the 6 columns whose names are listed in **cols**? <br>

**Answer**: We are essentially subsetting from the original **stationStatus** df; in particular, we are telling R to update **stationStatus** by assigning to it a new value, namely a subset of **stationStatus** which contains all original rows (I know that because the space to the left of the comma inside the square brackets (i.e. the row indicator) is empty which means we want all rows, but only the columns whose names are in the **cols** vector (I know that because the column indicator to the right of the [,] comma is **“cols”**.

```r
stationStatus <- stationStatus[,cols]
```

```r
stationStatus
```

Let’s create a couple of histograms:

```r
hist(stationStatus$num_docks_available)
hist(stationStatus$num_bikes_available)
#Plots the number of docks and bikes available
```

How about if we wanted to know how many stations in our data have at least one ebike available? 

```r
nrow(stationStatus[stationStatus$num_ebikes_available>0,])
```

Why did this work? **Answer**: Because we asked R to count the number of rows (aka, the number of records, where each record is a station) in the subset of **stationStatus** (I can tell it’s a subset by the [ ]) which has **all columns of stationStatus** (because we left the column indicator after the [,] comma empty) but **only those rows for which the num_ebikes_available column has a value greater than 0**.

What if in addition to knowing how many stations like that there are, we also wanted to save their subset to a new dataframe? The same logic applies:

```r
#Create a df of only the stations with at least one available ebike
ebike<-stationStatus[stationStatus$num_ebikes_available>0,]
```

```r
#Let’s take a look of the mean number of docks available in this new df
mean(ebike$num_docks_available)
```

We can even plot two values in the same visualization to create a scatterplot and study their relationship. 

```r
#Plot num of docks available against num of bikes available
plot(ebike$num_docks_available, ebike$num_bikes_available)
```
